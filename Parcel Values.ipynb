```python
import bs4, pandas as pd
from urllib.request import urlopen

# Creating a main dataframe
columns = ['Parcel', 'Tax Year', 'Class', 'Land', 'Building', 'Total']
main_df = pd.DataFrame(columns=columns)

# Test sample of parcels
parcels =  ['16 085 02 053', '16 139 01 033', '16 118 01 099']

# Looping through parcels to scrape data
for parcel in parcels:
    
    # Replacing the ' ' in the parcel with '%20' to avoid URL issues
    url_parcel = parcel.replace(' ', '%20')
    
    # Building URL and using BeautifulSoup to parse HTML for table values
    url = 'https://propertyappraisal.dekalbcountyga.gov/datalets/datalet.aspx?mode=value_history_main&UseSearch=no&pin=' + url_parcel
    html = urlopen(url).read()
    bs = bs4.BeautifulSoup(html, 'html.parser')
    table = bs.find(lambda tag: tag.name == 'table' and tag.has_attr('id') and tag['id'] == 'Appraised Values')
    rows = table.findAll(lambda tag: tag.name == 'tr')
    
    # Pull cell values for each row and organize into embedded lists
    value = []
    for tr in rows:
        td = tr.find_all('td')
        row = [tr.text for tr in td]
        value.append(row)
    
    # Create Pandas Dataframe
    df = pd.DataFrame(value[1:len(value)], columns=value[0])
    df['Parcel'] = parcel
    main_df = main_df.append(df)
    
main_df[main_df['Total'].notnull()]
```
